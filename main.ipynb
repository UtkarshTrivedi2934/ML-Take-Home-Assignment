```python
!pip install fastdtw
import torch
import torchaudio
import numpy as np
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw

def load_audio(path, sr=22050):
    wav, sample_rate = torchaudio.load(path)
    if sample_rate != sr:
        wav = torchaudio.functional.resample(wav, orig_freq=sample_rate, new_freq=sr)
    return wav


def l1_loss(gt, pred):
    return torch.nn.functional.l1_loss(gt, pred).item()

def l2_loss(gt, pred):
    return torch.nn.functional.mse_loss(gt, pred).item()


```

```python


# Load GT and predicted audio
gt = load_audio("/content/MeloTTS/melo/basetts_outputs_package_from_S3/en/EN-Default/speed_1.0/sent_000.wav")
pred = load_audio("/content/EN-default/output.wav")


# Compute Losses
print("Sample 1:")
print("  L1 Loss (wave):", l1_loss(gt, pred))
print("  L2 Loss (wave):", l2_loss(gt, pred))
```
